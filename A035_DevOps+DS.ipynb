```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

data = load_wine(as_frame=True)
df = data.frame

df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="target", data=df)
plt.title("Target Class Distribution (Wine Classes)")
plt.xlabel("Wine Class")
plt.ylabel("Count")
plt.show()
```

```python
df.iloc[:, :5].hist(figsize=(12,8))
plt.suptitle("Feature Distributions (First 5 Wine Features)")
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap="coolwarm", linewidths=0.2)
plt.title("Feature Correlation Heatmap (Wine Dataset)")
plt.show()
```
